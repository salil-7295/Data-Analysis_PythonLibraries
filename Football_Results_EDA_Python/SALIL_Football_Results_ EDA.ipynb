Lets Make it to TOP 10 Elite Tournments / Matches 


```python
Elite_Tournaments = ['FIFA World Cup','USA Cup','Copa América','UEFA Euro','Confederations Cup','Nations Cup','CCCF Championship','AFC Asian Cup','Gold Cup','African Cup of Nations']

Elite_Match = new_data.loc[new_data['tournament'].isin(Elite_Tournaments),:]
Elite_Match.sample(100)

# Elite Matches data frame will have only the Important tournament Results and Details
# This is also a process of reducing the dataset to the features that is important or interests more
```

## What will be  Summary of Total Goals per Tournament ( Home as well as Away 

#### Here we will make a new dataframe with away-score , home-score and total goals as per Selected 10 Tournaments

```python
Elite_Goal_Avg=Elite_Match.pivot_table(['home_score','away_score'],columns='tournament',aggfunc=np.sum)
Elite_Goal_Avg.loc['Total']=Elite_Goal_Avg.sum(axis=0)
Elite_Goal_Avg = Elite_Goal_Avg.T
Elite_Goal_Avg
```

## Visualize the Total Number of Goals per Tounaments

```python
plt.figure(figsize=(20,5))
sns.barplot(x=Elite_Goal_Avg.index,y=Elite_Goal_Avg['Total'])
```

#### Conclusion - Most number of gaols are in Copa America and FIFA WorldCup as per from the selected 10 Tournaments

## What will be Summary of Average number of Goals Per Tournament 

```python
num_matches =Elite_Match['tournament'].value_counts()
num_matches=num_matches.sort_index() # Get the count of matchs under specified tournament 
Elite_Goal_Avg['Num_Matches']=num_matches
Elite_Goal_Avg['Goal_Average']=Elite_Goal_Avg['Total']/Elite_Goal_Avg['Num_Matches'] # Calculating Average
Elite_Goal_Avg # Average
```

## Visualize the Average Goals Per Tournament Graphically 

```python
plt.figure(figsize=(20,5))
plt.title("Average Goals Per Important Tournaments")

ax = sns.barplot(x=Elite_Goal_Avg.index, y="Goal_Average", data=Elite_Goal_Avg,
                 color="Brown", saturation=.5)
```

### Conclusion - Almost all the selected Tournaments have the average Goal ranging 2 - 3 except CCFC Championship averaging above 4 goals per match
